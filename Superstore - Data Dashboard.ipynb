# Exploratory Data Analysis of Superstore Sales

### Data import and viewing

```python
import pandas as pd
df = pd.read_excel('Sample - Superstore.xls')
df.head()
```

```python
from lxml import html
import requests
headers= {"User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.82 Safari/537.36"}
url = "https://www23.statcan.gc.ca/imdb/p3VD.pl?Function=getVD&TVD=53971"
result = requests.get(url, headers=headers)
tree = html.fromstring(result.content)
# print(result.text)
state_names_tree = './/table[1]/tbody/tr/td[1]'
state_codes_tree = './/table[1]/tbody/tr/td[3]'

state_names = [title.text for title in tree.findall(state_names_tree)]
state_codes = [title.text for title in tree.findall(state_codes_tree)]
# for i in zip(state_name_tree, state_codes_tree):
#     state_names.append(i[0].text())
#     state_codes.append(i[1].text())
    
print(state_names, state_codes)

```

```python
mapper = dict(zip(state_names, state_codes))
df['State_Code'] = df['State'].map(mapper)
df.head()
```

#### Adding additional column to check lag between order and shiping

```python
df['Order to Ship Days'] = df['Ship Date']-df['Order Date']

# Changed Order to Ship Days to dtype int
df['Order to Ship Days'] = df['Order to Ship Days'].dt.total_seconds().astype('int')/86200
df['Order to Ship Days'] = df['Order to Ship Days'].astype('int')
df.head()
```

```python
import plotly.express as px
# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig = px.histogram(df, x='Order to Ship Days', histfunc='count')
fig.update_layout(
    title='Order to Ship Days lag histogram', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    barmode='group', 
    paper_bgcolor='#FFFFFF'
)
fig.show(renderer="iframe")
```

```python
## Orders by Segment

import plotly.express as px
# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig = px.histogram(df, x='Segment', y='Order Date', histfunc='count')
fig.update_layout(
    title='Orders count by Segment', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    barmode='group', 
    paper_bgcolor='#FFFFFF'
)
fig.show(renderer="iframe")
```

```python
# Pivoted into df
tmp_df = df[['State_Code', 'Quantity', 'Sales', 'Profit']].copy()
pivot_table = tmp_df.pivot_table(
    index=['State_Code'],
    values=['Sales', 'Quantity', 'Profit'],
    aggfunc={'Sales': ['sum'], 'Quantity': ['sum'], 'Profit': ['sum']}
)

pivot_table.columns = pivot_table.columns.get_level_values(0)
df_pivot = pivot_table.reset_index()
df_pivot
```

```python
## Orders Count by State

temp_df = df_pivot[['State_Code', 'Sales']].copy()
temp_df = temp_df.nlargest(10, columns=['Sales'])

import plotly.express as px
# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig = px.bar(temp_df, x='State_Code', y='Sales')
fig.update_layout(
    title='Top 10 States by Sales', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    barmode='group', 
    paper_bgcolor='#FFFFFF'
)
fig.show(renderer="iframe")
```

```python
df_temp = df_pivot.copy()
df_temp = df_temp.sort_values(['State_Code', 'Sales'])
df_temp = df_temp.groupby(['State_Code'])['Sales'].sum().reset_index()
# df_temp = df_temp.rename(columns = {'Quantity': 'Count'})
df_temp = df_temp.sort_values(['Sales'])
color_max = df_temp['State_Code'].max()
fig = px.choropleth(
    locations=df_temp['State_Code'], 
    color = df_temp['Sales'].astype(float), 
    locationmode="USA-states", 
    scope="usa", 
    color_continuous_scale = 'ylorrd', 
    labels={'color':'Sales'},
    title = 'Number of Simulated Locations by State',
#     animation_frame=df_temp["Sales"],
#     animation_group=df_temp["State"],
     range_color = [0, color_max]
    )
# fig.update_layout(
#     title_x =0.5,
#     title_font_size=30)
fig.show(renderer="iframe")
```

## Checking monthly Sales over the period of time

```python
# Pivoted into df
tmp_df = df[['Sales', 'Ship Date']].copy()
tmp_df['Ship Date (year-month)'] = tmp_df['Ship Date'].dt.strftime("%Y-%m")
pivot_table = tmp_df.pivot_table(
    index=['Ship Date (year-month)'],
    values=['Sales'],
    aggfunc={'Sales': ['count']}
)
# pivot_table = pivot_table.set_axis([flatten_column_header(col) for col in pivot_table.keys()], axis=1)
pivot_table.columns = pivot_table.columns.get_level_values(0)
df_pivot = pivot_table.reset_index()
df_pivot
```

#### Plotting periodic sales on bar chart

```python
import plotly.express as px
# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig = px.bar(df_pivot, x='Ship Date (year-month)', y='Sales', color='Sales')
fig.update_layout(
    title='Monthly Sales over the Period', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    barmode='group', 
    paper_bgcolor='#FFFFFF'
)
fig.show(renderer="iframe")
```

## Checking Sales over the different month of Year (Jan, Feb etc.)

```python
# Pivoted into df
tmp_df = df[['Profit', 'Sales', 'Order Date', 'Discount', 'Quantity']].copy()
tmp_df['Order Date (month)'] = tmp_df['Order Date'].dt.month
pivot_table2 = tmp_df.pivot_table(
    index=['Order Date (month)'],
    values=['Sales', 'Quantity', 'Profit', 'Discount'],
    aggfunc={'Sales': ['sum'], 'Quantity': ['sum'], 'Profit': ['sum'], 'Discount': ['mean']}
)
pivot_table2.columns = pivot_table2.columns.get_level_values(0)
df_pivot2 = pivot_table2.reset_index()
df_pivot2
```

#### Plotting sales by Month on bar chart

```python
import plotly.express as px
# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig2 = px.bar(df_pivot2, x='Order Date (month)', y='Sales', color='Profit')
fig2.update_layout(
    title='Sales by Month', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    barmode='group', 
    paper_bgcolor='#FFFFFF'
)
fig2.show(renderer="iframe")
```
