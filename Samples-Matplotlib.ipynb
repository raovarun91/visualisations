```python
from collections import Counter

lang_count = Counter()

for lang in lang_response:
    lang_count.update(lang.split(';'))
print(lang_count)
```

```python
language = []
popularity = []

for item in lang_count.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
    
print(language)
print(popularity)
```

```python
#Horizontal bar chart
plt.barh(language, popularity)
plt.xlabel('Language')
plt.ylabel('User Count')
plt.title("Most popular languages")

# plt.legend() #to show label values as legend
# plt.xticks(ticks=x_index, labels=ages_x) #As numpy generated index are labeled 1,2,3,4-- we need this workaround to show correct labels in front of ticks
# plt.grid(True) #to show grid lines

plt.tight_layout() 
# plt.savefig('Plots/Basic-seaborn-w-tight_layout.png')

plt.show()
```
